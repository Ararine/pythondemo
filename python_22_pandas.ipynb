#### 요일이 'Sun'이거나 'Sat'의 tip과 day의 값을 가져오기


```python
# print(data[data['day'] == 'Sun'][['tip','day']])
# print(data[data['day'] == 'Sat'][['tip','day']])
#dir(type(data[data['day']== 'Sat']))
# a = data[data['day'] == 'Sun']
# b = data[data['day'] == 'Sat']
# a.append(b)[['tip','day']]
data[(data['day'] == 'Sun') | (data['day'] == 'Sat')][['tip','day']]
a = data[(data['day'] == 'Sun') | (data['day'] == 'Sat')]
a['day'].unique() # 해당되는 값을 확인할 수 있다.
a[['tip','day']]
```

### describe
- describe는 기초적인 통계 분석을 해준다.

```python
# describe()의 기본은 데이터가 숫자로 된 컬럼만 가져온다.
data.describe()
```

```python
type(data.describe())
```

```python
# pandas는 numpy기반이라서 numpy 기능을 사용할 수 있다.
data.describe().T
```

```python
# describe()은 기본적으로 숫자형 데이터만 분석한다.
# include() 옵션을 사용하면 다른 분석할 데이터 타입을 지정할 수 있다.
data.describe(include=['object']) # object만 출력
data.describe(include=['float64', 'object']) # 숫자와 object 모두 출력
```

```python
# 데이터 타입이 object일 때만 가져온다.
#data.describe(include=['object'])
data.describe(include='O') # 대문자 알파벳 O, object와 같은 결과
```

```python
# 모든 타입의 데이터를 분석할 때는 include='all'로 설정한다.
data.describe(include='all')
```

```python
### where
```

```python
a = np.array([10,20,30,40,50])
a
```

```python
# where의 인자는 조건이다.
# 조건만 넣을 경우 True의 위치를 리턴한다.
np.where(a>30) # 출력되는 값은 index이다
```

```python
# where는 인자 개수에 따라서 기능이 달라진다.
# 조건이 True이면 0, False이면 100을 반환한다.
np.where(a>30, 0, 100) # (조건, True, False)
```

```python
## 기본
import numpy as np # numpy 패키지 가져오기
import pandas as pd # pandas 패키지 가져오기
import matplotlib.pyplot as plt # 시각화 패키지 가져오기

## 데이터 전처리
from konlpy.tag import Okt
from collections import Counter
from wordcloud import WordCloud
```

### 데이터 가져오기 

```python
df = pd.read_csv('data_crawling/movie_reviews.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

### 명사단어 추출 

#### 1. 세종사전 실행

```python
from konlpy.tag import Okt
twitter = Okt() # 세종사전 실행
```

#### 2. 단어 리스트 만들기

```python
word_list = []
word_list = df['story']
```

```python
word_list
```

#### 3. 형태소 분리 및 품사태깅

```python
sentences_tag = []

for sentence in word_list:
  morph = twitter.pos(sentence) # 형태소 태깅
  sentences_tag.append(morph)

print(morph)
```

#### 4. 명사만 추출 

```python
noun_list = []
for sentence in sentences_tag:
  for word, tag in sentence:
    if tag in ['Noun']:
      noun_list.append(word)
```

```python
noun_list
```

#### 5. 명사가 두 문자 이상인 것만 추출 

```python
len(noun_list)
noun_list = [n for n in noun_list if len(n) > 1]
noun_list[:10]
```

#### 6. 단어 카운트

```python
counts = Counter(noun_list)
tags = counts.most_common(30)
print(len(tags))
tags
```

### WordCloud 생성 

```python
wordcloud = WordCloud(font_path='NanumBarunGothic', background_color='white', width=800, height=600)
cloud = wordcloud.generate_from_frequencies(dict(tags))
plt.figure(figsize=(10,8))
plt.axis('off')
plt.imshow(cloud)
plt.show()
```
